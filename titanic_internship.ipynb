```python
import pandas as pd

# Load the Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Display the first few rows of the dataset
df.head()
```

```python
# Check for missing values
missing_values = df.isnull().sum()

# Fill missing values in 'Age' with the median value
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill missing values in 'Embarked' with the mode value
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop the 'Cabin' column as it has too many missing values
df.drop(columns=['Cabin'], inplace=True)

# Check for and remove duplicates
df.drop_duplicates(inplace=True)

# Display summary of the cleaned dataset
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Descriptive statistics
descriptive_stats = df.describe()

# Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Distribution of Fare
plt.figure(figsize=(10, 6))
sns.histplot(df['Fare'], kde=True, bins=30)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()

# Count plot for categorical variables
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Count')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', data=df)
plt.title('Gender Count')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='Embarked', data=df)
plt.title('Embarkation Port Count')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.show()

# Relationship between survival and age
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Survival vs Age')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()

# Relationship between survival and fare
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Survival vs Fare')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.show()

# Relationship between survival and passenger class
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival vs Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()

# Relationship between survival and gender
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival vs Gender')
plt.xlabel('Sex')
plt.ylabel('Count')
plt
```
